```
x = np.array([[1., 2., 3.], [4., 5., 6.]])
y = np.array([[6., 23.], [-1, 7], [8, 9]])
print(x)
print(y)
```

```
print(x.shape)
print(y.shape)
```

```
np.dot(x,y)
```

```
x.dot(y)     # x.dot(y) is equivalent to np.dot(x, y):
```

```
# A matrix product between a two-dimensional array and a suitably sized 
# onedimensional array results in a one-dimensional array:
one=np.ones(3)
print(one)
print(x)
np.dot(x, one)    # X is 2d while 'one' is 1d.
```

```
x @ np.ones(3)       # The @ symbol (as of Python 3.5) also works as an "infix operator" that performs matrix
                     # multiplication:    # One matrix is must for @ operator to perform action.
```

```
np.ones(6) @ np.ones(9)
```
