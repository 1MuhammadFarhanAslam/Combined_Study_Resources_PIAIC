```python
# variable_name[row]            # indexing
# variable_name[row,column]     #   indexing
# variable_name[start_row_index:end_row_index]   # only slicing of a row
# variable_name[start_row_index:end_row_index,start_column_index:end_column_index]   #Slicing of both rows and columns
```

```python
# 1  #when we use a single num it's means that we are talking about indexing
# 2  #when we use two index seprated by a comma its mean we are also
     #talking about indexing.
    
# 3  #when we use colan(:) sign its mean we are talking about slicing.

# 4  #when we use colon(:)sign providing index before comma,it;s mean we are
     #talking about slicing of only rows.

# 5  #when we use colon(:)sign providing index before and after comma(,) it;s mean we are
     #talking about slicing of both rows and columns.
```
